```python
print("請先輸入本學期的學分上限\n再輸入本學期的所有\n(1)課程名稱(限英文)\n\
(2)上課日(ex:mon、tue、wen、thu、fri)\n(3)上課起始時間(範圍:8~20)\n(4)上課結束時間(範圍:9~21)\n\
(5)此門課的學分\n若已輸入完畢請於\"課程名稱\"輸入0")#使用說明的文字敘述
item=1    
workday=["mon","tue","wen","thu","fri"] #上課日的範圍
mon=["","","","","","","","","","","","","","","","",""]   #因為後面程式碼需要有足夠的元素
tue=["","","","","","","","","","","","","","","","",""]
wen=["","","","","","","","","","","","","","","","",""]
thu=["","","","","","","","","","","","","","","","",""]
fri=["","","","","","","","","","","","","","","","",""]
totalcourse=[]  #用來收集所有課程
totalcredit=[]  #用來收集所有學分
while(1):       #利用迴圈避免使用者輸入錯誤的學分上限
    maxcredit=input("本學期的學分上限:")
    if maxcredit.isdecimal()and "0"<maxcredit<="31":
        break
while(1):
    course=input("第"+str(item)+"門課的課程名稱:")
    if course!="0":     #當輸入0則結束輸入
        day=input(course+"的上課日:")
        while day in workday:       #判斷是否所輸入的上課日在範圍內，若無則不會進入後續迴圈
            totalcourse.append(course)  #將所輸入的課程收集到totalcourse的空串列裡
            time1=input(course+"的上課起始時間:")
            time2=input(course+"的上課結束時間:")
            for a in range (0,int(time2)-int(time1)):  
                if day=="mon":
                    mon.insert(a+int(time1)-8,course)
                if day=="tue":
                    tue.insert(a+int(time1)-8,course)
                if day=="wen":
                    wen.insert(a+int(time1)-8,course)
                if day=="thu":
                    thu.insert(a+int(time1)-8,course)
                if day=="fri":
                    fri.insert(a+int(time1)-8,course)
                """以上是因為通常一門課會跨足多個時段，
                因此利用for迴圈重複加入課程進該上課日的串列裡面"""
            credit=input(course+"為幾學分:")        
            totalcredit.append(int(credit))     #將學分收集到totalcredit的空串列裡
            if 8<=int(time1)<=20 and 9<=int(time2)<=21 and int(time1)<int(time2): 
                item+=1         #判斷時間所輸入的開始及結束時間是否合理
                break
    else:   #若輸入0後便會結束整個迴圈
        print("已輸入完畢")
        break
title=["時段","時間","一","二","三","四","五"]
period=["1","2","3","4","C","D","5","6","7","8","E","F","G"]
with open("模擬課表.txt","w+",  encoding='utf-8') as file1:
        file1.write("時段,時間,一,二,三,四,五\n")
for j in range(0,13):
    aa=period[j]+",{0:02d}:00-{1:02d}:00,".format(j+8,j+9)+mon[j]+","+tue[j]+","+wen[j]+","+thu[j]+","+fri[j]
    with open("模擬課表.txt","a+",  encoding='utf-8') as file1:
        file1.write(aa+"\n")
for item in totalcourse:
    finalcourse=list(set(totalcourse))
print("所選的",len(finalcourse),"門課中包含","、".join(finalcourse),sep="")
if sum(totalcredit)<=int(maxcredit):
    print("您已選",sum(totalcredit),"學分",sep="")
else:
    print("您已選",sum(totalcredit),"學分，但超過學分上限",sum(totalcredit)-int(maxcredit),"學分",sep="")
file1=open('模擬課表.txt', 'r', encoding='utf-8')
list_lines=file1.readlines() #把檔案內所有的內容讀進一個串列中
file1.close()
for line in list_lines:
    print(line, sep=',')

```
