### 匯入基本的數據分析套件

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('https://quality.data.gov.tw/dq_download_csv.php?nid=20592&md5_url=e1accb446ecf527a2ae3bfde55a7ce40')
```

### 查看資料庫的內容

## 將資料進行分類-以受傷部位分類

```python
df.groupby(['受傷部位'])[[ "受傷部位次數"]].sum()
```

```python
heart_part=df.groupby(['受傷部位'])[[ "受傷部位次數"]].sum()
```

###  前5大受傷部位

```python
heart_part.sort_values(by=['受傷部位次數'], ascending=False).head()
```

## 將資料進行分類-以行業類別分類

```python
df.groupby(['行業別'])[[ "受傷部位次數"]].sum()
```

```python
heart_work=df.groupby(['行業別'])[[ "受傷部位次數"]].sum()
```

### 前5大容易受傷行業

```python
heart_work.sort_values(by=['受傷部位次數'], ascending=False).head()
```

## 結論
### 以行業別來看，在運輸及倉儲業受傷比例最高，且以受傷部位來看，手部的受傷機率最高，因此建議在上列高機率受傷的產業（運輸及倉儲業、住宿及餐飲業、電子零組件製造業、金屬製品製造業、批發及零售業），雇主需要特別注意員工發生職業傷害的狀況，尤其可以指、手、足、腿、頭這幾個部位為優先注意的對象。
