# HW4－記帳系統V4（可以畫出收支圓餅圖）

```python
import datetime
import matplotlib.pyplot as plt
initial_money=input("請問您目前擁有多少錢?")
income=[]
expense=[]
correctinput=["0","I","E"]
correct_kind_input=["1","2","3","4","5"]
income_item={}
expense_item={}
kind_dic={"1":"食","2":"衣","3":"住","4":"行","5":"樂"}
total_money=int(initial_money)
food_income=[]
clothing_income=[]
housing_income=[]
transportation_income=[]
entertainment_income=[]
food_expense=[]
clothing_expense=[]
housing_expense=[]
transportation_expense=[]
entertainment_expense=[]
while 1:
    income_or_expense=input("您要登記收入還是支出?（欲結束請輸入0） △貼心小提醒：收入請用I，支出請用E :")
    ISOTIMEFORMAT = '%Y-%m-%d %H:%M:%S'
    theTime = datetime.datetime.now().strftime(ISOTIMEFORMAT)
    if income_or_expense not in correctinput:
        print("您輸入的型式錯誤QQ請重新輸入")
    else:
        if income_or_expense=="0":
            for i in income:
                total_money+=int(i)
            for j in expense:
                total_money-=int(j)
            print("您目前帳戶擁有:",total_money,"元",sep="")
            break
        elif income_or_expense=="I"or"E":
            while 1:
                kind=input("1.食\n2.衣\n3.住\n4.行\n5.樂\n請輸入種類編號：")
                if kind in correct_kind_input:
                    item=input("請輸入金額：")
                    if income_or_expense=="I":                       
                        income.append(item)
                        income_item[theTime+kind_dic[kind]]=(item+"元")
                        if kind=="1":
                            food_income.append(item)
                        if kind=="2":
                            clothing_income.append(item)
                        if kind=="3":
                            housing_income.append(item)
                        if kind=="4":
                            transportation_income.append(item)
                        if kind=="5":
                            entertainment_income.append(item)
                        break
                    elif income_or_expense=="E":
                        expense_item[theTime+kind_dic[kind]]=(item+"元")
                        expense.append(item)
                        if kind=="1":
                            food_expense.append(item)
                        if kind=="2":
                            clothing_expense.append(item)
                        if kind=="3":
                            housing_expense.append(item)
                        if kind=="4":
                            transportation_expense.append(item)
                        if kind=="5":
                            entertainment_expense.append(item)
                        break
                else:
                    print("輸入錯誤!請重新輸入QQ")
def search_function(x):
    if x=="Y":
        a=0
        print("收入細項為:")
        for i in income_item:
            a+=1
            print(a,".",i,"->",income[a-1],"元",sep="")
        b=0
        print("支出細項為:")
        for j in expense_item:
            b+=1
            print(b,".",j,"->",expense[b-1],"元",sep="")
        print("謝謝您的使用!")
    if x=="N":
        print("謝謝您的使用!")
while 1:
    YN=input("是否想查詢收支細項?（請以Y或N輸入）")
    if YN=="Y"or"N":
        search_function(YN)
        break
    else:
        print("輸入錯誤!請重新輸入")
c=0
for i in [food_income,clothing_income,housing_income,transportation_income,entertainment_income]:
    for j in range(len(i)):
        c+=int(i[j])
        i.append(c)
        c=0
labels = ["Food","Clothing","Housing","Transportation","Entertainment"]
income_list=[food_income,clothing_income,housing_income,transportation_income,entertainment_income]
income=[]
income_new_labels=[]
def income_pie_chart():
    for i in income_list:
        if i!=[]:
            income.append(i[-1])
            income_new_labels.append(labels[income_list.index(i)])
    plt.pie(income,labels = income_new_labels,autopct='%1.1f%%')
    plt.title('Income Analysis')
    plt.show()
d=0
for i in [food_expense,clothing_expense,housing_expense,transportation_expense,entertainment_expense]:
    for j in range(len(i)):
        d+=int(i[j])
        i.append(d)
        d=0
expense_list=[food_expense,clothing_expense,housing_expense,transportation_expense,entertainment_expense]
expense=[]
expense_new_labels=[]

def expense_pie_chart():
    for i in expense_list:
        if i!=[]:
            expense.append(i[-1])
            expense_new_labels.append(labels[expense_list.index(i)])
    plt.pie(expense,labels = expense_new_labels,autopct='%1.1f%%')
    plt.title('Expense Analysis')
    plt.show()
income_pie_chart()
expense_pie_chart()
```
